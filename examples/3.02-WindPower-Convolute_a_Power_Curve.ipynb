# Power Curve Convolution

* Power curves only offer a "suggestion" of what a wind turbine should produce for a given wind speed
* In reality, many stochastics effects are affect this output
* RESkit allows for the creation of a convoluted power curve which takes stochastic responses into account
  - Begins with any general power curve
  - Convolves a guassian distribution with a wind-speed-dependent standard deviation onto the power curve

```python
from reskit import windpower
```

```python
# Power curve from a given capacity (kW) and rotor diameter (m)
pc = windpower.SyntheticPowerCurve(capacity=3000, rotordiam=140)

pc
```

```python
# Power curve from a specific capacity (in W/m2)
#  * As an example, standard deviation should be calculated with: std = ws*0.05 + 0.2
 
pc2 = windpower.convolutePowerCurveByGuassian(powerCurve=pc, stdScaling=0.05, stdBase=0.2)

pc2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(12,5))
plt.rc("font", size=18)

pc = windpower.SyntheticPowerCurve(capacity=3000, rotordiam=140)
plt.plot(pc.ws, pc.cf, label="Raw Power Curve", zorder=3, c="k", ls="--", lw=2)

pc2 = windpower.convolutePowerCurveByGuassian(powerCurve=pc, stdScaling=0.00, stdBase=0.2)
plt.plot(pc2.ws, pc2.cf, label="std=0.2*ws+0.00")

pc2 = windpower.convolutePowerCurveByGuassian(powerCurve=pc, stdScaling=0.00, stdBase=2.0)
plt.plot(pc2.ws, pc2.cf, label="std=2.0*ws+0.00")

pc2 = windpower.convolutePowerCurveByGuassian(powerCurve=pc, stdScaling=0.05, stdBase=0.2)
plt.plot(pc2.ws, pc2.cf, label="std=0.2*ws+0.05")

plt.legend()
plt.grid()
plt.ylabel("Capacity Factor")
plt.ylabel("Wind Speed [m/s]")

plt.show()
```
