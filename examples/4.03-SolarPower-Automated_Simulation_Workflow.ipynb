# Solar PV Simulation Workflow

```python
import reskit as rk
from reskit import solarpower

%matplotlib inline
```

---
## Single Location

```python
# Constants
LOC = (6.0,50.5)   # Given in longitude, latitude
ELEV = 100         # Elevation given in meters
SOURCE = rk._TEST_DATA_['weather_data'] # Path to climate model weather data (MERRA2)
```

```python
out = solarpower.workflowOpenFieldFixed(LOC, SOURCE, ELEV,
                                  module='1Soltech 1STH-345-WH', # Module to use for simulation
                                  azimuth=160,                   # Azimuth of system [degrees]
                                  tilt=44,                       # Tilt of system [degrees]
                                  capacity=7500,                 # Nameplate capacity [kW]
                                  extract='totalProduction'
                                 )

out.head(24)
```

```python
out.plot()
```

## Multiple Locations
- It is recommended to simulate multipl locations at once!

```python
# Constants
LOCS = [(6.0,50.5),
       (6.1,50.5),
       (6.2,50.5),
       (6.3,50.5),]

ELEVS = [100,100,200,300]

# tilts could be defined for each location as well...
# azimuths could be defined for each location as well...
# Name plate capacities could be defined for each location as well...
```

```python
out = solarpower.workflowOpenFieldFixed(LOCS, SOURCE, ELEVS,
                                  module='1Soltech 1STH-345-WH', # Module to use for simulation
                                  azimuth=160,                   # Azimuth of system [degrees]
                                  tilt=44,                       # Tilt of system [degrees]
                                  capacity=7500,                 # Nameplate capacity [kW]
                                  extract='raw'
                                 )

out.head(24)
```

```python
out.plot()
```

## GHI values can be scaled to match to high-resolution spatial trends
- Basically the same as the long-run average adjustment used in wind turbine simulations
- Comparisons are made against GHI values found in the Global Solar Atlas

```python
out = solarpower.workflowOpenFieldFixed(LOCS, SOURCE, ELEVS,
                                  module='1Soltech 1STH-345-WH', # Module to use for simulation
                                  azimuth=160,                   # Azimuth of system [degrees]
                                  tilt=44,                       # Tilt of system [degrees]
                                  capacity=7500,                 # Nameplate capacity [kW]
                                  extract='raw',
                                  ghiScaling=rk._TEST_DATA_['gsa-ghi-like.tif']
                                 )

out.head(24)
```

```python
out.plot()
```
