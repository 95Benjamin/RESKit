# Synthetic Power Curve

* RESKit allows synthetic power curves to be created in case they are not known

```python
from reskit import windpower
```

```python
# Power curve from a given capacity (kW) and rotor diameter (m)
pc = windpower.SyntheticPowerCurve(capacity=3000, rotordiam=140)

pc
```

```python
# Power curve from a specific capacity (in W/m2)
pc = windpower.SyntheticPowerCurve(specificCapacity=250)

pc
```

```python
# Adjustable cutoff wind speed
pc = windpower.SyntheticPowerCurve(specificCapacity=250, cutout=30)

pc
```

```python
# Direct access to the power curve's capacity factor values
#  - used 'pc.ws' for wind speed values
#  - used 'pc.cf' for capacity factor values

for i,ws,cf in zip(range(10), pc.ws, pc.cf):
    print("The capacity factor at {:.3f} m/s is {:.3f}".format(ws, cf))
    
```
